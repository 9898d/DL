```python
import transformers
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, GenerationConfig

tokenizer = AutoTokenizer.from_pretrained("BechirTrabelsi1/optLLM-small")
model = AutoModelForSeq2SeqLM.from_pretrained("BechirTrabelsi1/optLLM-small")
config = GenerationConfig(max_new_tokens=2000, do_sample=True, temperature=0.6)
```

```python
task = """We need to plan routes for 3 vehicles to deliver products to 10 customers.
Each vehicle has a capacity of 100 units. The customer demands are as follows: 10, 20, 15, 10, 5, 25, 30, 10, 15, 20 units.
The coordinates of the delivery locations are: (2,3), (5,5), (8,8), (1,2), (3,6), (7,2), (6,6), (5,9), (2,8), (4,4).
The warehouse is located at (0,0)."""
inputs = tokenizer(task, return_tensors="pt")
output_seq = model.generate(inputs["input_ids"], generation_config=config)
output = tokenizer.decode(output_seq[0], skip_special_tokens=True)
print(output)
```

```python
import re
pattern = r"'([^']+)'\s*:\s*(.*?)(?=,\s*'|$)"
matches = re.findall(pattern, output)
for match in matches:
    print(f"{match[0]}: {match[1]}")
```
